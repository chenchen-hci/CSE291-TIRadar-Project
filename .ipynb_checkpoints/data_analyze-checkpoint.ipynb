```python
import os
from analyze import Analyze
import numpy as np
import matplotlib.pyplot as plt

def xyzreader(filename):
    f = open(filename)
    buf = f.readlines()
    data = []

    for i in range(len(buf)):
        l = buf[i]
        l = l.split(",")
        l = [float(e) for e in l]
        data.append(l)
    return np.array(data, dtype = np.float64)
```

```python
def get_file_dir_single_person(node, node_person_distance, node_node_distance):
    directory = "data"
    filename = "log_" + str(node_person_distance) + "ft_distance_node_distance_" + str(node_node_distance) + "ft__node_" + str(node) + ".xyz"
    return os.path.join(os.getcwd(), directory, filename)

def get_file_dir_double_people(node, node_person_distance, node_node_distance, persons_distance):
    directory = "data_2_people"
    filename = "log_" + str(node_person_distance) + "ft_node_distance_" + str(node_node_distance) + "ft_people_dis_" + str(persons_distance) + "ft_node_" + str(node) + ".xyz"
    return os.path.join(os.getcwd(), directory, filename)

def serialize_result(result):
    # u, sigma, ratio
    data = []
    for i in range(len(result)):
        data.append([result[i][0][0], result[i][0][1], result[i][0][2], \
                    result[i][1][0][0], result[i][1][0][1], result[i][1][0][2], \
                    result[i][1][1][0], result[i][1][1][1], result[i][1][1][2], \
                    result[i][1][2][0], result[i][1][2][1], result[i][1][2][2], \
                    result[i][2]])
    return np.array(data, dtype = np.float64)
```

## node-person distance

```python
# evaluate single node case

# change node-person distance, keep node-node distance as 2ft
node1_result = []
node2_result = []

distances = [4, 6, 8, 12]

for d in distances:  
    d1 = xyzreader(get_file_dir_single_person(1, d, 2))
    d2 = xyzreader(get_file_dir_single_person(2, d, 2))

    d1 = Analyze(d1)
    d2 = Analyze(d2)

    u1, sigma1 = d1.compute_statistics()
    n1_clusters, n1_noise, cluster1, ratio1 = d1.clustering()
    u2, sigma2 = d1.compute_statistics()
    n2_clusters, n2_noise, cluster2, ratio2 = d2.clustering()
    node1_result.append([u1, sigma1, ratio1])
    node2_result.append([u2, sigma2, ratio2])

node1_result = serialize_result(node1_result)
node2_result = serialize_result(node2_result)
```

```python

d1 = xyzreader(get_file_dir_single_person(1, 12, 2))
d2 = xyzreader(get_file_dir_single_person(2, 12, 2))


d1 = Analyze(d1)
d2 = Analyze(d2)

u1, sigma1, g1 = d2.fit_gaussian(10, 0, 1)
u2, sigma2, g2 = d2.fit_gaussian(10, 0, 2)
u3, sigma3, g3 = d2.fit_gaussian(10, 1, 2)

print(u1, sigma1)
print(u2, sigma2)
print(u3, sigma3)

def sanitizer(x, y, z, threshold):
    new_x = []
    new_y = []
    new_z = []
    N = len(x)
    for i in range(N):
        if x[i] < - threshold or x[i] > threshold \
            or y[i] < -threshold or y[i] > threshold \
            or z[i] > threshold or z[i] < -threshold:
            continue
        else:
            new_x.append(x[i])
            new_y.append(y[i])
            new_z.append(z[i])

    return (np.array(new_x), np.array(new_y), np.array(new_z))

d1 = xyzreader(get_file_dir_single_person(1, 12, 2))
d2 = xyzreader(get_file_dir_single_person(2, 12, 2))

d1x = d1[:, 0]
d1y = d1[:, 1]
d1z = d1[:, 2]
d1x, d1y, d1z = sanitizer(d1x, d1y, d1z, 10)

d2x = d2[:, 0]
d2y = d2[:, 1]
d2z = d2[:, 2]
d2x, d2y, d2z = sanitizer(d2x, d2y, d2z, 10)

fig = plt.figure(6)
ax = fig.add_subplot(111, projection='3d')

deltad = 2 * 0.3048

ax.scatter(d1x, d1y, d1z, c = 'r', marker = 'o', label = "node1")
print(np.cov(np.vstack((np.array(d1x), np.array(d1y), np.array(d1z)))))

ax.scatter(d2x + deltad, d2y, d2z, c = 'b', marker = '*', label = "node2")
plt.legend(loc='lower right')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("scatter plot")

a1 = Analyze(d2)
n_cluster, n_noise, cluster, ratio = a1.clustering()
print(n_cluster)
```
